```python
import matplotlib.pyplot as plt
import pandas as pd
from pylab import *
mpl.rcParams['font.sans-serif'] = ['SimHei']


# 获取所有唯一的线路名称
unique_lines = data1['line_'].unique()

# 创建图形和轴
plt.figure(figsize=(10, 10), dpi=100)

# 遍历每条线路
for line_name in unique_lines:
    # 筛选出当前线路的行
    current_line = data1[data1['line_'] == line_name]
    
    # 提取经度值到列表x
    x = current_line['longitude'].tolist()
    
    # 提取纬度值到列表y
    y = current_line['latitude'].tolist()
    
    # 在图中绘制当前线路的折线图
    plt.plot(x, y, label=line_name, linestyle='-')
    plt.scatter(x,y,alpha=0.5,label="")

# 显示图例
plt.legend()

# 显示图形
plt.show()

##绘制整个北京市区地铁的总体图像
```
