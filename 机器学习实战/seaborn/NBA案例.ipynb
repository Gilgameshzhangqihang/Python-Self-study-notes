# 衍生变量的可视化实践

```python
def age_cut(df):
    if df.AGE <=24:
        return "young"
    elif df.AGE >=30:
        return "old"
    else:
        return "best"
```

```python
data=pd.read_csv("nba_2017_nba_players_with_salary.csv")
data
```

```python
##年龄划分
data["age_cut"]=data.apply(lambda x: age_cut(x), axis=1)
data.head()
```

```python
##方便计数
data["cut"]=1
```

```python
data.loc[data.age_cut=="old"].SALARY_MILLIONS
```

```python
##基于年龄对球员薪水和效率进行分析
sns.set_style("darkgrid")
plt.figure(figsize=(10,10))
plt.title("RPM and SALARY")
x1=data.loc[data.age_cut=="best"].SALARY_MILLIONS
y1=data.loc[data.age_cut=="best"].RPM
plt.plot(x1,y1,"^")

x2=data.loc[data.age_cut=="old"].SALARY_MILLIONS
y2=data.loc[data.age_cut=="old"].RPM
plt.plot(x2,y2,".")

x3=data.loc[data.age_cut=="young"].SALARY_MILLIONS
y3=data.loc[data.age_cut=="young"].RPM
plt.plot(x3,y3,"*")

plt.xlabel("SALARY_MILLIONS")
plt.ylabel("RPM")
```

```python
multi_data2=data.loc[:,['RPM','POINTS','TRB','AST','STL','BLK','age_cut']]
sns.pairplot(multi_data2,hue="age_cut")
```

# 球队的数据分析

```python
##按照球队的薪资水平排名
data_team=data.groupby(by="TEAM").agg({"SALARY_MILLIONS":np.mean})
data_team.sort_values(by="SALARY_MILLIONS",ascending=False).head(10)
```

```python
#按照分球队分年龄段，上榜球员降序排列
data_rpm=data.groupby(by=["TEAM","age_cut"]).agg({"SALARY_MILLIONS":np.mean,"RPM":np.mean,"PLAYER":np.size})
data_rpm.head()
```

```python
data_rpm.sort_values(by=["PLAYER","RPM"],ascending=False).head()
```
