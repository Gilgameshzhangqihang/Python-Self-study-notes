```python
import numpy as np
import seaborn as sns
np.random.seed(0)#确定随机数生成器生成的数据是一样的
array=np.random.randn(100)
ax=sns.distplot(array,bins=10,hist=False,kde=True,rug=False)
```

```python
array_random=np.random.randint(0,100,500)
sns.distplot(array_random,hist=False,rug=True)
```

```python
##绘制双变量分布
import numpy as np 
import pandas as pd
import seaborn as sns
df=pd.DataFrame({"x":np.random.randn(500),"y":np.random.randn(500)})
pd.set_option('display.max_columns', None)
# 显示所有行
pd.set_option('display.max_rows', None)
df
```

```python
sns.jointplot(x='x',y='y',data=df,kind="kde",color='r')
```

```python
##绘制二维直方图
sns.jointplot(x='x',y='y',data=df,kind="hex",color='r')
```
