# #创建多个子图

```python
import random
import matplotlib.pyplot as plt

x = range(60)
y_shanghai = [random.uniform(15, 18) for i in x]
y_beijing = [random.uniform(1, 3) for i in x]

figure, axes = plt.subplots(nrows=2, ncols=1,figsize=(20,8),dpi=80)

x_label=["11点{}分".format(i) for i in x]

axes[0].plot(x, y_shanghai, label="上海")
axes[0].set_title("上海温度变化")
axes[0].set_xticks(range(0, 60, 5))
axes[0].set_xticklabels(x_label[ : : 5])

axes[0].set_yticks(range(0, 40, 5))
axes[0].grid(linestyle='--', alpha=0.9)
axes[0].legend()

axes[1].plot(x, y_beijing, color="r", linestyle="--", label="北京")
axes[1].set_title("北京温度变化")
axes[1].set_xticks(range(0, 60, 5))

axes[1].set_yticks(range(0, 40, 5))
axes[1].grid(linestyle='--', alpha=0.9)
axes[1].legend()

plt.show()

```

# 直方图histogram

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

time=np.random.randint(90,150,size=60)
plt.rcParams['font.family'] = 'SimSun'
plt.figure(figsize=(20,8),dpi=80)

distance=2
group_num=int((max(time)-min(time))/distance)
plt.hist(time,bins=group_num,density=True)##density是y轴变不变成频率的意思

plt.xticks(range(min(time),max(time),distance))
plt.grid(linestyle='--',alpha=0.5)
plt.title("绘制的直方图")

plt.show()
```
